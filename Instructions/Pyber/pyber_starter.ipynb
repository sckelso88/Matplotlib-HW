```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_ride_data = pd.merge(ride_data, city_data, how="left", on= ["city", "city"])

city_ride_data
```

```python
# Obtain the x and y coordinates for each of the three city types
urban_cities = city_ride_data[city_ride_data["type"]== "Urban"]
urban_ride_count = urban_cities.groupby(["city"]).count()["ride_id"]
urban_avg_fair = urban_cities.groupby(["city"]).mean()["fare"]
urban_driver_count = urban_cities.groupby(["city"]).mean()["driver_count"]
#urban_driver_count
suburban_cities = city_ride_data[city_ride_data["type"]== "Suburban"]
suburban_ride_count = suburban_cities.groupby(["city"]).count()["ride_id"]
suburban_avg_fair = suburban_cities.groupby(["city"]).mean()["fare"]
suburban_driver_count = suburban_cities.groupby(["city"]).mean()["driver_count"]
#suburban_cities
rural_cities = city_ride_data[city_ride_data["type"]== "Rural"]
rural_ride_count = rural_cities.groupby(["city"]).count()["ride_id"]
rural_avg_fair = rural_cities.groupby(["city"]).mean()["fare"]
rural_driver_count = rural_cities.groupby(["city"]).mean()["driver_count"]
rural_cities
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.scatter(urban_ride_count,
            urban_avg_fair, 
            s=10*urban_driver_count, 
            c="blue", edgecolor="black", linewidth=1, marker="o", alpha=0.7,
            label= "Urban")
plt.scatter(suburban_ride_count,
            suburban_avg_fair, 
            s=10*suburban_driver_count, 
            c="green", edgecolor="black", linewidth=1, marker="o", alpha=0.7,
            label= "Suburban")
plt.scatter(rural_ride_count,
            rural_avg_fair, 
            s=10*rural_driver_count, 
            c="orange", edgecolor="black", linewidth=1, marker="o", alpha=0.7,
            label= "Rural")

# Incorporate the other graph properties
plt.title ("Pyber Rider Sharing Data")
plt.ylabel("Average Fare")
plt.xlabel("Total Number of Rides")
# Create a legend
plt.legend(title="City Types")
plt.show()
# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("fig1.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_percents = 100* city_ride_data.groupby(["type"]).sum()["fare"]/city_ride_data["fare"].sum()
type_percents
# Build Pie Chart
plt.pie(type_percents,
       labels=["Rural", "Suburban", "Urban"], autopct='%1.1f%%')
plt.title("% of Total Fares by City Type")
plt.show()
# Save Figure
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_city_type = 100* city_ride_data.groupby(["type"]).count()["ride_id"]/city_ride_data["ride_id"].count()
rides_city_type
# Build Pie Chart
plt.pie(rides_city_type,
       labels=["Rural", "Suburban", "Urban"], autopct='%1.1f%%')
plt.title("Total Rides by City Type")
plt.show()
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_percent_city_type = 100* city_data.groupby(["type"]).sum()["driver_count"]/city_data["driver_count"].sum()
#drivers_percent_city_type
plt.pie(drivers_percent_city_type,
       labels=["Rural", "Suburban", "Urban"], autopct='%1.1f%%')
plt.title("Total Drivers by City Type")
plt.show()
# Build Pie Charts

# Save Figure
plt.savefig("fig5.png")
```
